```python
# Import the dependencies.
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from citipy import citipy
```

```python
# generate 2k random lat and lngs and zip them
lats = np.random.uniform(low=-90.000, high=90.000, size=2000)
lngs = np.random.uniform(low=-180.000, high=180.000, size=2000)
lat_lngs = zip(lats, lngs)
lat_lngs # checked
```

```python
# adding lat and lngs into a list
coordinates = list(lat_lngs)
# create an list for random cities.
cities = []
# Identify the nearest city for each latitude and longitude combination.
for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name

    # If the city is unique, then we will add it to the cities list.
    if city not in cities:
        cities.append(city)
# Print the city count to confirm sufficient count.
len(cities) # found 748 cities from citipy
```

```python
print(city)
```

```python
# perform api call 
import requests
from config import weather_api_key

#  url for opeanweathermap api Call.
url = "http://api.openweathermap.org/data/2.5/weather?units=Metric&APPID=" + weather_api_key
```

```python
# loop over cities to retrieve data
city_data = []

print("beginning data retrieval")
print("--------------------------------")

no_info_city = []
for city in cities:
    # create an endpoint URL for each city.
    city_url = url + "&q=" + city.replace(" ", "+")
    
    
    try:
        # parse data in json
        city_info = requests.get(city_url).json()
        
        country = city_info['sys']['country']
        lng = city_info['coord']['lon']
        lat = city_info['coord']['lat']
        max_temp = city_info['main']['temp_max']
        humidity = city_info['main']['humidity']
        cloudiness = city_info['clouds']['all']
        wind_speed = city_info['wind']['speed']
        c_description = city_info['weather'][0]['description']
        
        city_data.append({"City":city.title(),
                          "Country": country,
                          "Lat":lat,
                          "Lng": lng,
                          "Max Temp": max_temp,
                          "Humidity": humidity,
                          "Cloudiness": cloudiness,
                          "Wind Speed": wind_speed,
                          "Current Description": c_description})
        
    except:
        print(f"{city} not found. Skiping")
        pass
print("--------------------------------")
print("Retrieval complete")
print("--------------------------------")
```

```python
len(city_data) #  found 687 cities' weather info, out of 748 cities. 
```

```python
# add results into a new df
result_df = pd.DataFrame(city_data)
result_df.head()
```

```python
# formatting lat and lng columns
result_df["Lat"] = result_df["Lat"].map("{:.2f}".format)
result_df["Lng"] = result_df["Lng"].map("{:.2f}".format)
result_df.head(10)
```

```python
# export result_df to csv
output_csv = "WeatherPy_Database.csv"
result_df.to_csv(output_csv)
```
